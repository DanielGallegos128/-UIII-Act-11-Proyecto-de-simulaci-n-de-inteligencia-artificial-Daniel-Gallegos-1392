### Supervisor: Línea de producción, empleados bajo mando, reporte de incidencias, turno de trabajo, indicadores de calidad y planta.

```python
# --------------------------------------------------------
# PROGRAMA: SISTEMA DE GESTIÓN DE SUPERVISORES
# REGLA: Los corchetes [] (Singular = La Lista)
#        Las llaves    {} (Plural   = Los Atributos)
# --------------------------------------------------------

# 1. CORCHETES [] -> SINGULAR: Representa a "La Lista" única de la fábrica.
lista_supervisor = []


# 2. FUNCIÓN: Agregar un nuevo supervisor
def agregar_supervisor():
    print("\n--- REGISTRAR NUEVO SUPERVISOR ---")

    # Entrada de datos (Key, Value)
    planta = input("Planta (Ej. Norte, Sur): ")
    linea = input("Línea de producción: ")
    empleados = input("Cantidad de empleados bajo su mando: ")
    turno = input("Turno de trabajo (Matutino/Vespertino/Nocturno): ")
    incidencias = input("Reporte de incidencias de hoy: ")
    calidad = input("Porcentaje de indicadores de calidad (Ej. 95%): ")

    # LLAVES {} -> PLURAL: Representa a "Los Atributos" (muchos datos de un supervisor)
    atributos_supervisores = {
        "Planta": planta,
        "Línea": linea,
        "Empleados": empleados,
        "Turno": turno,
        "Incidencias": incidencias,
        "Calidad": calidad,
    }

    # Guardamos los Atributos {} dentro de La Lista []
    lista_supervisor.append(atributos_supervisores)
    print("\n✅ ¡Supervisor registrado con éxito!")


# 3. FUNCIÓN: Mostrar los datos en forma de tabla alineada
def mostrar_supervisores():
    print("\n--- REPORTE DE SUPERVISORES EN PLANTA ---")

    # Usamos IF para evaluar si La Lista [] (singular) está vacía
    if len(lista_supervisor) == 0:
        print("No hay supervisores registrados en el sistema todavía.")
    else:
        # Encabezados de la tabla perfectamente alineados
        print(
            f"{'Planta':<12} | {'Línea':<10} | {'Empleados':<10} | {'Turno':<12} | {'Calidad':<10} | {'Incidencias'}"
        )
        print("-" * 80)

        # Ciclo FOR para revisar los Atributos {} (plural) dentro de La Lista [] (singular)
        for atributos in lista_supervisor:
            print(
                f"{atributos['Planta']:<12} | {atributos['Línea']:<10} | {atributos['Empleados']:<10} | {atributos['Turno']:<12} | {atributos['Calidad']:<10} | {atributos['Incidencias']}"
            )


# 4. FUNCIÓN: Simulación de Inteligencia Artificial
def asistente_ia():
    print("\n--- ASISTENTE DE IA: DIAGNÓSTICO DE PLANTA ---")

    if len(lista_supervisor) == 0:
        print("La IA no tiene datos para analizar. Registra un supervisor primero.")
    else:
        print("Analizando las líneas de producción...\n")

        # La IA analiza cada grupo de Atributos {} usando IF, ELIF y ELSE
        for atributos in lista_supervisor:
            print(
                f"** Análisis para Línea {atributos['Línea']} ({atributos['Planta']}):"
            )

            # Limpieza rápida del texto para evaluar el número de calidad
            porcentaje_calidad = int(atributos["Calidad"].replace("%", ""))

            if (
                atributos["Incidencias"].lower() != "ninguna"
                and atributos["Incidencias"].lower() != "ninguno"
                and atributos["Incidencias"] != "0"
            ):
                print(
                    f"   ⚠ ALERTA: Se reporta la incidencia '{atributos['Incidencias']}'. Se sugiere enviar soporte técnico."
                )
            elif porcentaje_calidad < 85:
                print(
                    f"   📉 Alerta de Calidad: El rendimiento es de {atributos['Calidad']}. Requiere auditoría."
                )
            elif int(atributos["Empleados"]) > 20:
                print(
                    "   👥 Nota de Personal: Alta concentración de operadores. El supervisor podría requerir ayuda."
                )
            else:
                print(
                    "   🟢 Estado: Operación estable. Todo corre bajo los parámetros normales."
                )
        print("\n--- Fin del reporte de IA ---")


# 5. MENÚ INTERACTIVO PRINCIPAL (Ciclo infinito)
while True:
    print("\n========================================")
    print("  SISTEMA DE CONTROL DE PRODUCCIÓN INDUSTRIAL")
    print("========================================")
    print("1. Registrar nuevo Supervisor")
    print("2. Ver tabla de Supervisores")
    print("3. Ejecutar Asistente de IA (Diagnóstico)")
    print("4. Salir del sistema")

    opcion = input("\nSelecciona una opción (1-4): ")

    if opcion == "1":
        agregar_supervisor()
    elif opcion == "2":
        mostrar_supervisores()
    elif opcion == "3":
        asistente_ia()
    elif opcion == "4":
        print("\nCerrando sistema... ¡Buen turno de trabajo!")
        break
    else:
        print("\n❌ Opción no válida. Por favor, elige un número del 1 al 4.")
```
